```python
# Import pustaka yang dibutuhkan 
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
from playsound import playsound
```

```python
# penggunaan pustaka untuk mendeteksi dan menganalisis tangan dalam citra atau video
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
#  implementasi kode
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    # membaca frame dari video selama webcam masih terbuka
    while cap.isOpened(): 

        # membaca frame dari video dan ret akan menjadi true kalo framenya berhasil terbaca
        ret, frame = cap.read()

        # BGR 2 RGB karena media pipe menggunakan format RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Memutar kamera agar menampilkan efek cermin 
        image = cv2.flip(image, 1)

        # Set flag
        image.flags.writeable = False

        # Detections gambar dan hasilnya disimpan ke dalam results 
        results = hands.process(image)

        # Set flag to true
        image.flags.writeable = True

        # RGB 2 BGR untuk ditampilkan lagi
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Menampilkan hasil deteksi tangan (landmarks tangan)
        print(results)

        # Rendering apakah tangan terdeteksi dalam frame, jika ada maka melakukan iterasi untuk tangan yang terdeteksi dan gambar landmark tangan
        # dan garis penghubungnya
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)

        # tampilkan hasil deteksi
        cv2.imshow('Hand Tracking', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```

```python
# Inisialisasi tangkapan video
cap = cv2.VideoCapture(0)

# Inisialisasi MediaPipe Hands
with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands:
    while cap.isOpened():
        ret, frame = cap.read()

        # BGR 2 RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

        # Flip secara horizontal
        image = cv2.flip(image, 1)

        # Setel flag
        image.flags.writeable = False

        # Deteksi
        results = hands.process(image)

        # Setel flag menjadi true
        image.flags.writeable = True

        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Deteksi
        print(results)

        # Tampilkan hasil
        if results.multi_hand_landmarks:
            for num, hand in enumerate(results.multi_hand_landmarks):
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS)

                # Dapatkan landmark
                landmarks = hand.landmark

                wrist = landmarks[0]
                
                # Ibu jari 
                thumb_cmc = landmarks[1]
                thumb_mcp = landmarks[2]
                thumb_pip = landmarks[3]
                thumb_tip = landmarks[4]

                # Telunjuk
                index_finger_mcp = landmarks[5]
                index_finger_pip = landmarks[6]
                index_finger_dip = landmarks[7]
                index_finger_tip = landmarks[8]


                # Jari tengah
                middle_finger_mcp = landmarks[9]
                middle_finger_pip = landmarks[10]
                middle_finger_dip = landmarks[11]
                middle_finger_tip = landmarks[12]


                # Jari manis 
                ring_finger_mcp = landmarks[13]
                ring_finger_pip = landmarks[14]
                ring_finger_dip = landmarks[15]
                ring_finger_tip = landmarks[16]

                # Kelingking
                pinky_mcp = landmarks[17]
                pinky_pip = landmarks[18]
                pinky_dip = landmarks[19]
                pinky_tip = landmarks[20]
                
                # Cetak nilai koordinat landmark

                # atas kecil bawah besar
                print(f"Index Finger mcp (Y): ({index_finger_mcp.y})")
                print(f"Index Finger tip (Y): ({index_finger_tip.y})")

                # kiri kecil kanan besar
                print(f"Index Finger mcp (X): ({index_finger_mcp.x})") 
                print(f"Index Finger tip (X): ({index_finger_tip.x})")
                
                # Bandingkan koordinat y untuk menentukan arah penunjukan
                # Tangga nada SI
                if (index_finger_tip.y < index_finger_mcp.y and thumb_tip.y < thumb_cmc.y and thumb_tip.x < index_finger_mcp.x 
                and thumb_tip.x < thumb_pip.x and middle_finger_tip.y > middle_finger_mcp.y and ring_finger_tip.y > ring_finger_mcp.y 
                and pinky_tip.y > pinky_mcp.y):
                    cv2.putText(image, "Tangga nada SI", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                    # playsound('/Kuliah/SEMESTER 5/Pengolahan Citra Digital/Praktek/TUBES/ngeteh.wav')
                
                # Tangga nada Do
                elif(thumb_tip.y > index_finger_pip.y and thumb_tip.x > index_finger_pip.x and index_finger_mcp.y < index_finger_pip.y
                and middle_finger_mcp.y < middle_finger_pip.y and ring_finger_mcp.y < ring_finger_pip.y
                and pinky_mcp.y < pinky_pip.y):
                    cv2.putText(image, "Tangga nada DO", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)

                # Tangga nada Re
                elif(thumb_tip.y < thumb_cmc.y and index_finger_tip.y < index_finger_mcp.y 
                and middle_finger_tip.y < middle_finger_mcp.y and ring_finger_tip.y < ring_finger_mcp.y
                and pinky_tip.y < pinky_mcp.y and thumb_tip.x < index_finger_mcp.x 
                and index_finger_tip.y < thumb_tip.y and middle_finger_tip.y < thumb_tip.y and ring_finger_tip.y < thumb_tip.y
                and pinky_tip.y < thumb_tip.y):
                    cv2.putText(image, "Tangga nada RE", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                    
                # Tangga nada Mi
                
                # Tangga nada Fa 
                elif(index_finger_tip.y > thumb_tip.y and thumb_tip.x < index_finger_tip.x and index_finger_tip.y > index_finger_pip.y
                and middle_finger_mcp.y < middle_finger_pip.y and ring_finger_mcp.y < ring_finger_pip.y
                and pinky_mcp.y < pinky_pip.y and index_finger_tip.y > wrist.y ):
                    cv2.putText(image, "Tangga nada FA", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                    
                # Tangga nada Sol 
                elif(thumb_tip.y < pinky_tip.y and thumb_tip.y < index_finger_tip.y):
                    cv2.putText(image, "Tangga nada SOL", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                    
                # Tangga nada La
                elif(index_finger_tip.y > thumb_tip.y and thumb_tip.x > index_finger_tip.x and index_finger_tip.y > index_finger_pip.y 
                and middle_finger_tip.y > middle_finger_pip.y and ring_finger_tip.y > ring_finger_pip.y
                and pinky_tip.y > pinky_pip.y):
                    cv2.putText(image, "Tangga nada La", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
                    
        # Tampilkan gambar
        cv2.imshow('Hand Tracking', image)

        # Periksa tombol keluar
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

# Bebaskan tangkapan dan tutup semua jendela
cap.release()
cv2.destroyAllWindows()
```
